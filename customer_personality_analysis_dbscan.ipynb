## Data Pre-Processing

```python
# Scale the features in the df

# Not sure if we want to drop these columns below
# del_cols = ['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1','AcceptedCmp2', 'Complain', 'Response']
# features = df_cleaned.drop(del_cols, axis=1)
# scaler = StandardScaler()
# scaler.fit(features)
# scaled_features = pd.DataFrame(scaler.transform(features),columns= features.columns)

scaler = StandardScaler()
scaler.fit(df_cleaned)
df_scaled = pd.DataFrame(scaler.transform(df_cleaned),columns= df_cleaned.columns)
```

```python
#PCA to reduce dimensions to 3
pca = PCA(n_components=3)
pca.fit(df_scaled)
PCA_ds = pd.DataFrame(pca.transform(df_scaled), columns=(["col1","col2", "col3"]))
PCA_ds.describe().T
```

```python
#3D projection of data
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(PCA_ds["col1"],PCA_ds["col2"],PCA_ds["col3"], marker="o",cmap = 'viridis')
ax.set_title("Data Reduced to 3 Dimensions")
plt.show()
```

```python
# Determine number of clusters with Elbow Method
elbow = KElbowVisualizer(KMeans(), k=10)
elbow.fit(PCA_ds)
elbow.show()
```

## Model: DBSCAN

### Finding optimal distance hyperparameter for DBSCAN

```python
# Plot Sorted k-distance Graph to visually find optimal epsilon distance for DBSCAN

from sklearn.neighbors import NearestNeighbors

# k = 2*dimension of data - 1
k = 2 * PCA_ds.shape[-1] - 1

# Create Nearest Neighbor Model
neigh = NearestNeighbors(n_neighbors=k)
# Fit model to the data
neigh.fit(PCA_ds)
# Get the distances of every data point to each of it's k nearest neighbors
distances, _ = neigh.kneighbors(PCA_ds)
# Sort the points based on distance
distances = np.sort(distances, axis=0)
# Take only the distance of the point to the k neighbor
k_distance = distances[:, k-1]

# Plot sorted k-distance graph where x axis are the number of points and y axis is the distance of the point to it's k neighbor
plt.figure(figsize=(8,8))
plt.plot(k_distance)
plt.xlabel('Points in the dataset')
plt.ylabel(f'Sorted {k}-nearest neighbor distance')
plt.grid(True, linestyle="--", color='black', alpha=0.4)

```

```python
# From the K-distance plot, epsilon set as 0.6 seems to be an optimal distance.
```

### DBSCAN model implementation

```python
from sklearn.cluster import DBSCAN
# Create DBSCAN Clustering model, min samples choose 2*dim, basically double amount of dimensions
DP = DBSCAN(eps=0.6, min_samples=2*PCA_ds.shape[-1])
# fit model to the transformed data, predict clusters
DP_df = DP.fit_predict(PCA_ds)
#Adding the predicted cluster to the original dataframe and PCA transformed dataframe
PCA_ds["Clusters_DBSCAN"] = DP_df
df_cleaned["Clusters_DBSCAN"] = DP_df
```

### DBSCAN Prediction Results

```python
#Plotting the clusters
fig = plt.figure(figsize=(10,8))
ax = plt.subplot(111, projection='3d', label="bla")
ax.scatter(PCA_ds["col1"],PCA_ds["col2"],PCA_ds["col3"], s=40, c=PCA_ds["Clusters_DBSCAN"], marker='o', cmap = 'viridis' )
ax.set_title("The Plot Of The Clusters")
plt.show()
```

```python
# Countplot of clusters
pl = sns.countplot(x=PCA_ds["Clusters_DBSCAN"])
pl.set_title("Distribution Of The Clusters")
plt.show()
```

### DBSCAN predicted clusters analysis

```python
# Analyze characteristics of each cluster part 1
for i in range(5):
    cluster_data = df_cleaned[DP_df == i]

    print(f"Cluster {i}: Size = {len(cluster_data)}")
    print("Mean of each feature:")
    print(cluster_data.mean(axis=0))
    print("-------------------------")
```

```python
# Analyze characteristics of each cluster part 2

# create an empty list
feature_list = []

# iterate over each feature
for column in df_cleaned.columns:
    if column != 'Clusters_DBSCAN':
        # get the mean value for each cluster
        mean_val = [cluster_data[column].mean() for cluster_data in [df_cleaned[DP_df == i] for i in range(5)]]

        # round the value into two decimal places
        mean_val_rounded = [round(value, 2) for value in mean_val]

        # create a dataframe
        df = {'Clusters_DBSCAN': [f'Clusters_DBSCAN {i}' for i in range(5)], f'Mean_{column}': mean_val_rounded}
        mean_df = pd.DataFrame(df)

        # append the dataframe to the list
        feature_list.append(mean_df)


for i, df in enumerate(feature_list):
    print(df)
    print("------------------------------------------")
```
